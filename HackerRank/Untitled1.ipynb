```python
def solution(A):
    A.sort()
    N = len(A)
    
    i = 0
    previous = 0
    while i < N:
        current = A[i]

        if current > 0:
            if current > previous + 1:
                return previous + 1
            else:
                previous = current
        
        i += 1
    
    return max(previous+1, current+1)
A=[-1,2,4,9,5]
solution(A)
```
